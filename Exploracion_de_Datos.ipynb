```python
duplicadas = melbourne_df.T.duplicated()

# Mostrar las columnas duplicadas (si existen)
columnas_duplicadas = melbourne_df.columns[duplicadas].tolist()

if columnas_duplicadas:
    print("Las siguientes columnas son duplicadas:", columnas_duplicadas)
else:
    print("No hay columnas duplicadas en el dataset.")
```

```python
# Filtrar columnas duplicadas cuyo nombre finaliza con un espacio en blanco
columnas_con_espacio = list(filter(lambda col: col.endswith(" "), columnas_duplicadas))

# Mostrar las columnas identificadas
print("Columnas a eliminar:", columnas_con_espacio)
```

```python
# Contar valores nulos estándar (NaN o pd.NA)
nulos_estandar = melbourne_df.isnull().sum()

# Contar valores sospechosos que podrían representar nulos
valores_sospechosos = ["?", "NA", "N/A", "-", "None", "", " "]

nulos_no_estandar = {valor: (melbourne_df == valor).sum().sum() for valor in valores_sospechosos}

# Mostrar resultados
print("Valores nulos detectados por pandas:\n", nulos_estandar[nulos_estandar > 0])
print("Posibles valores nulos mal representados:")
for valor, cantidad in nulos_no_estandar.items():
    if cantidad > 0:
        print(f"'{valor}': {cantidad} ocurrencias")
```

## Convertir los tipos de datos

```python
cols_categoric = [
    "Type",
    "SellerG",
    "Regionname",
    "CouncilArea",
    "Method",
    "Suburb",
    "Address",
    "Postcode",
]

melbourne_df[cols_categoric] = melbourne_df[cols_categoric].astype("category")
```

```python
cols_numeric_float = ["Price", "BuildingArea", "Landsize", "Lattitude", "Longtitude", "Distance"]

melbourne_df[cols_numeric_float] = melbourne_df[cols_numeric_float].astype("float")
```

```python
cols_numeric_int = ["Car", "Propertycount", "Rooms", "Bathroom", "Bedroom2", "YearBuilt"]

melbourne_df[cols_numeric_int] = melbourne_df[cols_numeric_int].astype("Int64")
```

```python
cols_datetime = ["Date"]
melbourne_df[cols_datetime] = melbourne_df[cols_datetime].astype("datetime64[ns]")
```

```python
schema = pa.Table.from_pandas(melbourne_df).schema
```

## Guardar los datos

```python
melbourne_df = melbourne_df.reset_index(drop=True)
# Crear el directorio si no existe
(DATA_DIR / "02_intermediate").mkdir(parents=True, exist_ok=True)

# Guardar en formato Parquet
melbourne_df.to_parquet(DATA_DIR / "02_intermediate/melbourne_type_fixed.parquet", index=False)
```
